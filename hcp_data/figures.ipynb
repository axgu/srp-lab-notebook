```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%store -r ff_acc
%store -r gru_acc
%store -r transformer_acc
%store -r logreg_acc
%store -r transformer_rand_acc
```

```python
xAx = [i for i in range(0,90)]
plt.plot(xAx, transformer_acc, label="transformer")
plt.plot(xAx, logreg_acc, label="log-reg", color="C3")
plt.plot(xAx, ff_acc, label="ff", color="C2")
plt.plot(xAx, gru_acc, label="gru", color="C1")
plt.plot(xAx, transformer_rand_acc, label="transformer-random", linestyle='dashed', color="C0")
plt.xlabel("Time (s)")
plt.ylabel("Accuracy")
plt.ylim(0,1)
plt.xlim(0,90)
plt.title("Time-varying Classification Accuracy")
plt.legend()
plt.show()
```

```python
%store -r gru_attention_acc
%store -r gru_rand_acc
xAx = [i for i in range(0,90)]
plt.plot(xAx, gru_attention_acc, label="gru-attention")
plt.plot(xAx, gru_acc, label="gru")
plt.plot(xAx, gru_rand_acc, label="permutation", linestyle="dashed", color="C0")
plt.xlabel("Time (s)")
plt.ylabel("Accuracy")
plt.ylim(0,1)
plt.xlim(0,90)
plt.title("Time-varying Classification Accuracy for GRU")
plt.legend()
plt.show()
```

```python
%store -r ff_attention_acc
%store -r ff_rand_acc
xAx = [i for i in range(0,90)]
plt.plot(xAx, ff_attention_acc, label="ff-attention", color="C0")
plt.plot(xAx, ff_acc, label="ff", color="C1")
plt.plot(xAx, ff_rand_acc, label="permutation", color="C0")
plt.xlabel("Time (s)")
plt.ylabel("Accuracy")
plt.ylim(0,1)
plt.xlim(0,90)
plt.title("Time-varying Classification Accuracy for FF")
plt.legend()
plt.show()
```
