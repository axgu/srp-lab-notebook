```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
xAx = [i for i in range(0,90)]
plt.plot(xAx, logreg_acc, label="log-reg", color="C3")
plt.plot(xAx, ff_acc, label="ff", color="C1")
plt.plot(xAx, gru_acc, label="gru", color="C2")
plt.plot(xAx, transformer_acc, label="transformer", color="C0")
plt.plot(xAx, transformer_rand_acc, label="random", color="C0", linestyle="dashed")
plt.xlabel("Time (s)")
plt.ylabel("Accuracy")
plt.ylim(0,1)
plt.xlim(0,90)
plt.title("Time-varying Classification Accuracy")
plt.legend()
plt.show()
```
