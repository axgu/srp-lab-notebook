```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
%store -r ff_acc
%store -r gru_acc
%store -r transformer_acc
```

```python
xAx = [i for i in range(0,90)]
plt.plot(xAx, ff_acc, label="ff")
plt.plot(xAx, gru_acc, label="gru")
plt.plot(xAx, transformer_acc, label="transformer")
plt.xlabel("Time (s)")
plt.ylabel("Accuracy")
plt.ylim(0,1)
plt.xlim(0,90)
plt.title("Time-varying Classification Accuracy")
plt.legend()
plt.show()
```
